```python

#Define a class to receive the characteristics of each line detection
class Line():
    def __init__(self):
        # Line was detected before.
        self.detected = False  
        #polynomial coefficients for the current fit
        self.current_fit = np.array([0,0,0], dtype='float')  
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float')
        # The last best fit.
        self.best_fit = np.array([0,0,0], dtype='float')
```

```python
def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    

# findcorners takes for argument a grayscale image and its number of rows and columns and returns a bool (ret) 
# and array (corners) 
def findcorners(img, nrow, ncol):
    gray = grayscale(img)
    return cv2.findChessboardCorners(gray, (nrow,ncol), None)

def CalibrateCamera(img, obj_list, img_list):
    gray = grayscale(img)
    return cv2.calibrateCamera(obj_list, img_list, gray.shape[::-1], None, None)
    
def Undistorted(img, distort_mat, distort_coeff):
    return cv2.undistort(img, distort_mat, distort_coeff, None, distort_mat)
    
def warper(img, src, dst):
    # Compute and apply perpective transform
    img_size = (img.shape[1], img.shape[0])
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_NEAREST)  # keep same size as input image
    return warped    

def gradient_binary(img, sx_min, sx_max):
    gray = grayscale(img)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1,0)
    abs_sobelx = np.absolute(sobelx)
    scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
    bin_sobelx= np.zeros_like(scaled_sobel)
    bin_sobelx[(scaled_sobel >= sx_min) &(scaled_sobel <= sx_max)] = 1
    return bin_sobelx

def color_binary(img, col_min, col_max):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]
    bin_color = np.zeros_like(s_channel)
    bin_color[(s_channel >= col_min) & (s_channel <= col_max)] = 1
    return bin_color

def combined_binary(img, sx_min, sx_max, col_min, col_max):
    
    gradient = gradient_binary(img, sx_min, sx_max)
    color = color_binary(img, col_min, col_max)
    
    # Combine the two binary thresholds
    combined_bin_img = np.zeros_like(gradient)
    combined_bin_img[(color == 1) | (gradient == 1)] = 1

    return combined_bin_img
    

def make_histo(img):
    
    histogram = np.sum(img[img.shape[0]//2:,:], axis =0)
    #plt.plot(histogram)
    #plt.savefig('./output_images/histogram.jpg')
    midpoint = np.int(histogram.shape[0]//2)
    
    right_start = np.argmax(histogram[midpoint:])+midpoint
    
    
    left_start = np.argmax(histogram[:midpoint])
    
    return right_start, left_start


def find_lanes_pixels(img):
    right_curr_pos, left_curr_pos = make_histo(img)
    out_img = np.dstack((img, img, img))*255
    #Windows Parameters
    nwindows = 10
    window_width = 100
    window_height = np.int(img.shape[0]//nwindows)
    minpix = 50
    
    nonzero = img.nonzero()
    nonzerox = np.array(nonzero[1])
    nonzeroy = np.array(nonzero[0])
    
    
    left_lane_pix = []
    right_lane_pix = []
    
    for window in range(nwindows):
        window_y_low = img.shape[0] - (window+1)*window_height
        window_y_high = img.shape[0] -  window*window_height
        window_leftx_low = left_curr_pos - window_width
        window_leftx_high = left_curr_pos + window_width
        window_rightx_low = right_curr_pos - window_width
        window_rightx_high = right_curr_pos + window_width
       
        
        cv2.rectangle(out_img,(window_leftx_low,window_y_low),(window_leftx_high,window_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(window_rightx_low,window_y_low),(window_rightx_high,window_y_high),(0,255,0), 2) 
        
        
        good_left_pix = ((nonzeroy > window_y_low) &
                         (nonzeroy <= window_y_high) & 
                         (nonzerox > window_leftx_low) &
                         (nonzerox <= window_leftx_high)).nonzero()[0]
        
        good_right_pix = ((nonzeroy > window_y_low) &
                         (nonzeroy <= window_y_high) & 
                         (nonzerox > window_rightx_low) &
                         (nonzerox <= window_rightx_high)).nonzero()[0]
        
        left_lane_pix.append(good_left_pix)
        right_lane_pix.append(good_right_pix)
        
        if (len (good_left_pix) > minpix):
            left_curr_pos = np.int(np.mean(nonzerox[good_left_pix]))
        if(len (good_right_pix) > minpix):
            right_curr_pos = np.int(np.mean(nonzerox[good_right_pix]))
    try:
        left_lane_pix = np.concatenate(left_lane_pix)
        right_lane_pix = np.concatenate(right_lane_pix)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass
    
    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_pix]
    lefty = nonzeroy[left_lane_pix] 
    rightx = nonzerox[right_lane_pix]
    righty = nonzeroy[right_lane_pix]                     
                         
        
    return leftx, lefty, rightx, righty, out_img

def fitted_lanes(img, leftx, lefty, rightx, righty):
    
    left_fit_param_0 = []
    left_fit_param_1 = []
    left_fit_param_2 = []
    
    right_fit_param_0 = []
    right_fit_param_1 = []
    right_fit_param_2 = []
    
    left_fit_param = np.polyfit(lefty, leftx, 2)
    right_fit_param = np.polyfit(righty, rightx, 2)
    
    
    left_fit_param_0.append(left_fit_param[0])
    left_fit_param_1.append(left_fit_param[1])
    left_fit_param_2.append(left_fit_param[2])
    
    left_fit_param = [np.mean(left_fit_param_0), 
                      np.mean(left_fit_param_1), 
                      np.mean(left_fit_param_2)]
    
    left_fit_param = np.array(left_fit_param)
    
    right_fit_param_0.append(right_fit_param[0])
    right_fit_param_1.append(right_fit_param[1])
    right_fit_param_2.append(right_fit_param[2])
    
    right_fit_param = [np.mean(right_fit_param_0), 
                       np.mean(right_fit_param_1), 
                       np.mean(right_fit_param_2)]
    
    right_fit_param = np.array(right_fit_param)
    
    
    polyfunc = np.linspace(0, img.shape[0]-1, img.shape[0])
   
    try:
        fitted_left_lane = left_fit_param[0]*polyfunc**2 + left_fit_param[1]*polyfunc + left_fit_param[2]
        fitted_right_lane = right_fit_param[0]*polyfunc**2 + right_fit_param[1]*polyfunc + right_fit_param[2]
        
    except TypeError:
        # Avoids an error if `left_lane` and `right_lane` are still none or incorrect
        print('The function failed to fit a line!')
        fitted_left_lane = 1*polyfunc**2 + 1*polyfunc
        fitted_right_lane = 1*polyfunc**2 + 1*polyfunc
    
    return left_fit_param, right_fit_param, fitted_left_lane, fitted_right_lane, polyfunc

def visualized_fit_lanes(img, leftx, lefty, rightx, righty, fitted_left_lane, fitted_right_lane, polyfunc, out_img):
    # Draw the windows on the visualization image
    
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]
    return out_img
    


def extrapolated_lanes(img,left_fit, right_fit):
    nonzero = img.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    margin = 50
    
    left_lane_pix = ((nonzerox >= ((left_fit[0] * nonzeroy **2) + left_fit[1]* nonzeroy + left_fit[2] - margin))  
                     & (nonzerox < ((left_fit[0] * nonzeroy **2) + left_fit[1] * nonzeroy + left_fit[2] + margin)))
                    
    
    right_lane_pix = ((nonzerox >= ((right_fit[0] * nonzeroy **2) + right_fit[1] * nonzeroy + right_fit[2] - margin))
                      & (nonzerox < ((right_fit[0] * nonzeroy **2) + right_fit[1] * nonzeroy + right_fit[2]+ margin)))
                     
                     
                     
    leftx = nonzerox[left_lane_pix]
    lefty = nonzeroy[left_lane_pix] 
    rightx = nonzerox[right_lane_pix]
    righty = nonzeroy[right_lane_pix]
    
    
    return leftx, lefty, rightx, righty
    

def vis_lane_projection(image, warped, left_fitx, right_fitx, ploty, src_pts, dst_pts):
    # Create an image to draw the lines on
    warp_zero = np.zeros_like(warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    
    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    
    #Calculate the conversion Matrix.
    Minv = cv2.getPerspectiveTransform(dst_pts, src_pts)
    #Minv = cv2.getPerspectiveTransform(dst, src)
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (image.shape[1], image.shape[0])) 
    # Combine the result with the original image
    
    #undistort_im = Undistorted(image, distort_mat, distort_coeff)
    #result = cv2.addWeighted(undistort_im, 1, newwarp, 0.3, 0)
    
    
    return newwarp


def measure_curvature_pixels(img):
    '''
    Calculates the curvature of polynomial functions in pixels.
    '''
    # Start by generating our fake example data
    # Make sure to feed in your real data instead in your project!
    ploty, left_fit, right_fit = fitfunc(img)
    
    # Define y-value where we want radius of curvature
    # We'll choose the maximum y-value, corresponding to the bottom of the image
    y_eval = np.max(ploty)
    
    # Calculation of R_curve (radius of curvature)
    left_curverad = ((1 + (2*left_fit[0]*y_eval + left_fit[1])**2)**1.5) / np.absolute(2*left_fit[0])
    right_curverad = ((1 + (2*right_fit[0]*y_eval + right_fit[1])**2)**1.5) / np.absolute(2*right_fit[0])
    
    return left_curverad, right_curverad


# Calculate the radius of curvature in pixels for both lane lines
#left_curverad, right_curverad = measure_curvature_pixels(img)

#print(left_curverad, right_curverad)
# Should see values of 1625.06 and 1976.30 here, if using
# the default `generate_data` function with given seed number



```

```python
def measure_curvature_real(warped, left_fitx, right_fitx, ploty):
    
    
    
    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    
    y_eval = np.max(ploty)
    
    # Fit a second order polynomial to pixel positions in each lane line
    left_fit_cr = np.polyfit(ploty * ym_per_pix, left_fitx * xm_per_pix, 2)
    right_fit_cr = np.polyfit(ploty * ym_per_pix, right_fitx * xm_per_pix, 2)
    
    
    # Calculation of radius of curvature.
    left_curvature = ((1 + (2* left_fit_cr[0] * y_eval * ym_per_pix + left_fit_cr[1])**2)**1.5) / (np.abs(2*left_fit_cr[0]))  
    right_curvature = ((1 + (2* right_fit_cr[0] * y_eval * ym_per_pix + right_fit_cr[1])**2)**1.5) / (np.abs(2*right_fit_cr[0]))
    
    return left_curvature, right_curvature
    
```

```python

def car_offset(left_fitx, right_fitx, image):
    ## Image mid horizontal position 
    ym_per_pix = 30/720 
    xm_per_pix = 3.7/700
    mid_imgx = image.shape[1]//2
        
    ## Car position with respect to the lane
    car_pos = (left_fitx[-1] + right_fitx[-1])/2
    
    ## Horizontal car offset 
    offset = (mid_imgx - car_pos) * xm_per_pix
    print

    return offset
```

```python
def Display_curvature(image, left_fitx, right_fitx, ploty):    
    
    curvature = measure_curvature_real(image, left_fitx, right_fitx, ploty)
    
    offset = car_offset(left_fitx, right_fitx, image)

    # Display lane curvature
    out_image = image.copy()
    cv2.putText(out_image, 'Left lane line curvature: {:.2f} m'.format(curvature[0]), 
                (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)
    cv2.putText(out_image, 'Right lane line curvature: {:.2f} m'.format(curvature[1]), 
                (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)
    
    # Display car offset
    cv2.putText(out_image, 'car offset: {:.2f} m'.format(offset), 
                (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)
    
    return out_image
```

```python
def sanity_check(left_fit_param, right_fit_param):
    left_lane.diffs = np.abs(left_lane.current_fit - left_fit_param)
    right_lane.diffs = np.abs(right_lane.current_fit - right_fit_param)
    
    
    
    
    if ((left_lane.diffs[0] > 0.001) or (left_lane.diffs[1] > 1) or (left_lane.diffs[2] > 50)):
        
        left_lane.detected = False
        left_lane.current_fit = np.array([0,0,0], dtype='float')
        
        print(left_lane.diffs)
    
    
    if ((right_lane.diffs[0] > 0.001)\
        or (right_lane.diffs[1] > 1)\
        or (right_lane.diffs[2] > 50)):
        right_lane.detected = False
        right_lane.current_fit = np.array([0,0,0], dtype='float') 
        
    return
    
```

```python
def warp_image(img, src_pts, dst_pts):
    #  Getting the image size
    img_size = (img.shape[1], img.shape[0])
    
    # Calculate the transformation matrix form src and dst point.
    M = cv2.getPerspectiveTransform(src_pts, dst_pts)
    
    # Generate the warped image (Bird view from normal view, or vice versa)
    warped = cv2.warpPerspective(img, M, img_size)
    
    return warped
```

```python
def process_video(clip):
    undist = Undistorted(clip, mtx, dist)
    
    src, dst = transfpoints(clip)
    
    warped_img = warp_image(undist, src, dst)
    
    filtered_im = combined_binary(warped_img, 
                                low_gradient_tresh, 
                                high_gradient_tresh,
                                low_color_thresh, high_color_thresh)
    
    #filtered_im = cv2.resize(filtered_im1, (240, 135), interpolation=cv2.INTER_LINEAR)
    
    if((left_lane.detected==False) or (right_lane.detected==False)): 
        leftx, lefty, rightx, righty, out_img = find_lanes_pixels(filtered_im)
        left_fit_param, right_fit_param, fitted_left_lane, fitted_right_lane, polyfunc = fitted_lanes(filtered_im, leftx, lefty, rightx, righty)
           
        left_lane.detected=True
        right_lane.detected=True
        left_lane.current_fit = left_fit_param             
        right_lane.current_fit = right_fit_param  
    
    if((left_lane.detected==True) and (right_lane.detected==True)):
        leftx, lefty, rightx, righty = extrapolated_lanes(filtered_im,left_lane.current_fit,right_lane.current_fit)
       
                  
        if leftx == [] or lefty == [] or rightx == [] or righty == []:
            
            left_lane.detected=False
            right_lane.detected=False
            left_lane.current_fit = np.array([0,0,0], dtype='float')            
            right_lane.current_fit = np.array([0,0,0], dtype='float') 
            
        else:
            left_fit_param, right_fit_param, fitted_left_lane, fitted_right_lane, polyfunc = fitted_lanes(filtered_im, leftx, lefty, rightx, righty)
            sanity_check(left_fit_param, right_fit_param)
            
            newwarp = vis_lane_projection(undist, filtered_im, fitted_left_lane, fitted_right_lane, polyfunc, src, dst)
            result = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)
             
            curvature = measure_curvature_real(result, fitted_left_lane, fitted_right_lane, polyfunc)
            offset = car_offset(fitted_left_lane, fitted_right_lane, result)
            #result = Display_curvature(result,fitted_left_lane, fitted_right_lane, polyfunc)
           
            # Print the results
            # print('Left line curvature:', curvature[0], 'meters')
            # print('Right line curvature:', curvature[1], 'meters')
            # print(offset)
            if (curvature[0] >=2000.0 or curvature[1] >= 2000.0 ):
                cv2.putText(result, 'Straight Lanes', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)
            else:
                cv2.putText(result, 'Left lane line curvature: {:.2f} m'.format(curvature[0]), 
                            (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)
                cv2.putText(result, 'Right lane line curvature: {:.2f} m'.format(curvature[1]), 
                            (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)
            
                # Display car offset
                cv2.putText(result, 'car offset: {:.2f} m'.format(offset), 
                            (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 5)

    return result
```
